```python
import os
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np
import torch
from src.load_data import get_dataloaders_DRIVE
from src.load_data import get_dataloaders_PH2
import torchvision
```

```python
batch_size = 8
trainloader, valloader,_ = get_dataloaders_DRIVE(batch_size=batch_size, size=256, data_path="../data/DRIVE/training")
        
```

```python
# visualize a batch of images and tagets
def show_batch(img_batch, targets):
    fig = plt.figure(figsize=(16, 6))
    for i in range(8):
        ax = fig.add_subplot(2, 8, i + 1)
        ax.imshow(img_batch[i].permute(1, 2, 0))
        ax.axis("off")
    for i in range(8):
        ax = fig.add_subplot(2, 8, 8+i + 1)
        ax.imshow(targets[i])
        ax.axis("off")
    plt.show()



# get a batch of training data
inputs, targets, mask = next(iter(trainloader))

# show images
show_batch(inputs, targets)
```
